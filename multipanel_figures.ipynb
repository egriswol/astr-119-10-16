```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi, 100)
print(x[-1], 2*np.pi)

y = np.sin(x)
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)
```

```python
#call subplots to make a multipanel figure. 1 row, 2 columns in this example
f, axarr = plt.subplots(1,2)

#treat axarr as an array, from left to right

#first panel
axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')
#second panel
axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

#making space between the panels
f.subplots_adjust(wspace=1)

#resolution, 2 options
axarr[0].set_aspect('equal')
axarr[1].set_aspect(np.pi)
```

```python
#adjust figure size
fig = plt.figure(figsize=(6,6))

plt.plot(x, y, label='y = sin(x)')
plt.plot(x, z, label=r'$y = \cos(x)$')
plt.plot(x, w, label=r'$y = \sin(4x)$')
plt.plot(x, v, label=r'$y = \cos(4x)$')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.xlim([0, 2*np.pi])
plt.ylim([-1.2,1.2])
plt.legend(loc=1, framealpha=0.95) #makes a legend upper right corner, transparent

#fix axis ratio
plt.gca().set_aspect(np.pi/1.2)
```
